```
import numpy as np
from scipy import integrate

def f(x):
  return (x+1)/np.sqrt(3-x**2)

integrate.quad(f, -np.pi/4, np.pi/2)
```

```
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def dydx(y,x):
  return -2.5*y

x=np.linspace(0, 2, 100)
y0=0
y=odeint(dydx, y0, x)
y=np.array(y).flatten()
```

```
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_title=('Решение ОДУ 1-ого порядка')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

ax.plot(x,y,color='pink', linewidth=3)
lgnd=ax.legend(['y'], loc='upper right', shadow=True)
lgnd.get_frame().set_facecolor('#ffb19a')
plt.show()
```
