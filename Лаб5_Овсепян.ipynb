№ 5.2.2

Дан список температурных изменений в течение дня (целые числа). Известно, что измеряющее устройство иногда сбоит и записывает отсутствие температуры (значение None).

Выведите среднюю температуру за наблюдаемый промежуток времени, предварительно очистив список от неопределенных значений. Гарантируется, что хотя бы одно определенное значение в списке есть.

```
def avg(data):
    """Вернуть среднее значение 'data'.
    Параметры:
        data(list): список температурных изменений без None
    Результат:
        float: среднее значение
    """
    return sum(data) / len(data)
   

def cleared_data(data):
    """Вернуть список без значений None.
    Параметры:
        data(list): список температурных изменений 
    Результат:
        list: список температурных изменений без None
    """
    for i in data:
       if i is None: 
          data.remove(i)
    return data       

n = int(input('Кол-во измерений: '))
# Если с клавиатуры вводится прочерк, измерения нет
# (необходимо добавить в список None)
rawData = list()
for i in range(n):
    d = input('Измерение {}-е: '.format(i+1))
    rawData.append(None if d == '-' else float(d))   


# Получить очищенный список и среднее значение
clearedData = cleared_data(rawData)
print('Средняя температура: {0:.2f}'.format(avg(clearedData)))
# --------------
# Пример вывода:
#
# Кол-во измерений: 3
# Измерение 1-е: 10
# Измерение 2-е: -
# Измерение 3-е: 20
# Средняя температура: 15.00
```

№ 5.2.3

Выведите все счастливые номера билетов в диапазоне от a до b (положительные целые числа, a<b), если известно, что счастливым считается номер, у которого количество четных цифр равно количеству нечетных.

```
def is_lucky(num):
    """Является ли 'num' номером счастливого билета?
    Параметры:
        num(int): номер билета
    Результат:
        bool: True - да, False - нет.
    """
    num_even_count = 0
    num_odd_count = 0
    while num != 0:
        if num % 2 == 0:
           num_even_count += 1
        else:
           num_odd_count += 1  
        num = num // 10
    return True if num_even_count == num_odd_count else False  
def lucky_numbers(a, b):
    """Вернуть список счастливых номеров от 'a' до 'b'.
    Параметры:
        a(int): начало диапазона
        b(int): конец диапазона
    Результат:
        list: список счастливых номеров
    """
    num = list()
    for i in range(a, b + 1):
        if is_lucky(i):
           num.append(i) 
    return num  

a = int(input('Первый номер билета: '))
b = int(input('Последний номер билета: '))
# Вывод должен быть осуществлен в строку с одним пробелом-разделителем
num = lucky_numbers(a, b)
for i in num:
    print(i, end = ' ')
# --------------
# Пример вывода:
#
# Первый номер билета: 10
# Последний номер билета: 25
# 10 12 14 16 18 21 23 25
```

№ 5.2.4

Дата характеризуется тремя натуральными числами: день, месяц и год. Учитывая, что год может быть високосным, реализуйте две функции, которые определяют вчерашнюю и завтрашнюю дату.

```
def is_leap(year):
    """Является ли 'year' високосным годом?
    Параметры:
        year(int): год
    Результат:
        bool: True - да, False - нет.
    """
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False   
    elif year % 4 == 0:
        return True

def days(month, year):
    """Вернуть количество дней в месяце 'month' года 'year'.
    Параметры:
        month(int): номер месяца
        year(int): год
    Результат:
        int: количество дней 
    """
    if month in (1,3,5,7,8,10,12):
        return 31
    elif month == 2:
        if is_leap(year):
            return 29
        else:
            return 28    
    else: 
        return 30

def previous_date(day, month, year):
    """Вернуть день, месяц, год предыдущего дня.
    Параметры:
        day(int): день 
        month(int): месяц
        year(int): год
    Результат:
        int: день, месяц, год предыдущего дня.
    """
    if month == 1:
        month = 12
        year -= 1
    else:     
        month -= 1 
    if day == 1:
         day =  days(month, year)
    else:
         day -= 1    
    return day, month, year

def next_date(day, month, year):
    """Вернуть день, месяц, год следующего дня.
    Параметры:
        day(int): день 
        month(int): месяц
        year(int): год
    Результат:
        int: день, месяц, год следующего дня
    """
    if day == days(month, year):
        day = 1
        if month == 12:
            month = 1
            year += 1
        else:     
            month += 1 
    else:
        day += 1   
    return day, month, year      

date = input('День, месяц, год через пробел: ').split(sep=' ')
day = int(date[0]) 
month = int(date[1])
year = int(date[2])
pday, pmonth, pyear = previous_date(day, month, year)
print('Предыдущий день: {:0>2}/{:0>2}/{:0>4}'.format(pday, pmonth, pyear))
nday, nmonth, nyear = next_date(day, month, year)
print('Следующий день: {:0>2}/{:0>2}/{:0>4}'.format(nday, nmonth, nyear))


# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 3 2000
# Предыдущий день: 29/02/2000
# Следующий день: 02/03/2000
```

№ 5.2.7

Даны n предложений. Определите, сколько из них содержат хотя бы одну цифру.

```
def has_digits(sentence):
    """`Определить, есть ли в предложении цифры.
    Параметры:
        sentence(str): предложение
    Результат:
        bool: True - да, False - нет.
        '"""
    for i in sentence: 
        if i in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
            return True
            return False    
            

def sentences_with_digits_count(sentences):
    """Определить, сколько предложений содержат хотя бы одну цифру.
    Параметры:
        sentences(str): предложения содержащие хотя бы одну цифру
    Результат:
        int: количество предложений.
    """
    count = 0
    for i in sentences:
        if has_digits(i):
            count += 1 
    return count

n = int(input('Введите количество предложений: '))
sentences = list()
for i in range(0, n):
    sentence = input('Введите предложение № {}: '.format(i+1))
    sentences.append(sentence)
t = sentences_with_digits_count(sentences)
print('Предложений с цифрой = ', t)
# --------------
# Пример вывода:
#
# Введите количество предложений: 3
# Введите предложение №1:
# Просто текст
# Введите предложение №2:
# Текст с цифрой 1 (один)
# Введите предложение №3:
# Тут нет цифры
# Предложений с цифрой = 1
```

№ 5.2.9

Напишите функции для перевода числа (N∈[2;16]):

из 10-й системы счисления в N-ю;

из N-й системы счисления в 10-ю.

```
LETTERS_EX = {10: "A", 11: "B", 12: "C", 13: "D", 14: "E", 15: "F"}
DIGITS_EX = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}

def to_base(number, base):
    """Преобразовать десятичное число 'number' в с.с. 'base'.
    Параметры:
        number (int): десятичное число;
        base (int): система счисления.
    Результат:
        str: число в с.с. 'base'."""
    result = ''
    while number >= base:
        ost = number % base
        if ost > 10:
            result = LETTERS_EX[ost] + result
        else:
            result = str(ost) + result
        number = number // base
    if number > 10:
        result = LETTERS_EX[number] + result
    else:
        result = str(number) + result
    return result

def from_base(number, base):
    """Преобразовать  число 'number' из с.с. 'base' в десятичную.
    Параметры:
        number (str): число в с.с. 'base'
        base (int): начальная система счисления.
    Результат:
        int: число в 10-й с.с."""
    count = 0
    result = 0
    for letter in reversed(number):
        if letter in DIGITS_EX.keys():
            i = DIGITS_EX[letter]
        else: 
            i = int(letter)    
        result += i * (base ** count)
        count += 1
    return result
print('Перевод десятичного числа 689 в шестнадцатиричную с.с. ', to_base(689, 16))
print('Перевод числа 2B1 из шестнадцатиричной с.с. в десятичную ', from_base('2B1', 16))
# --------------
# Пример вывода:
#
# Нет
```

№ 5.2.10

Для введенного предложения выведите статистику символ=количество. Регистр букв не учитывается.



```

def sentence_stats(sentence):
    
    d = dict()
    sentence = sentence.lower()
    for i in sentence:
        d[i] = d.get(i, 0) + 1
    return d
s = input('Введите предложение: ')
print(sentence_stats(s))
# --------------
# Пример вывода:
#
# Введите предложение: мама МЫла РамУ
# {'л': 1, 'р': 1, 'у': 1, 'м': 4, 'а': 4, 'ы': 1, ' ': 2}
```
