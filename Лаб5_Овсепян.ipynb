```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_csv('bestsellers with categories.csv')
df
```

# **Box plots**

```
sns.boxplot(df['User Rating'])
```

```
sns.boxplot(x='User Rating', data=df)
```

**Kernel density estimate**

```
sns.kdeplot(df['User Rating'])
```

```
sns.kdeplot(df['Reviews'])
```

```
sns.kdeplot(df['Price'])
```

# **Violin plot**

```
sns.violinplot(x='User Rating', y='Price', data=df)
```

# **Count plots**

```
sns.countplot(df['User Rating'])
```

# **Heatmap**

```
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
```

# **Facet grid**

```
g=sns.FacetGrid(df, col='Genre', row='Year')
g.map(plt.hist,'User Rating')
```

# **Regplot**

```
sns.regplot(x='User Rating', y='Price', data=df)
```

# **lmplot**

```
sns.lmplot(x='Year', y='Price', row='User Rating', data=df, col='Genre')
```

# **Strip plot**

```
sns.stripplot(x='Genre', y='Price', data=df)
```

# **Swarm plot**

```
sns.swarmplot(x='Genre', y='Price', data=df)
```

```
sns.violinplot(x='Genre', y='Price', data=df, palette='pink')
sns.swarmplot(x='Genre', y='Price', data=df, color='blue', size=3)
```

# **Pair plot**

```
sns.pairplot(data=df)
```

```
sns.pairplot(df, hue='Genre', palette='Set1')
```

# **Joint plot**

```
sns.jointplot(x='Year', y='User Rating', data=df)
```
