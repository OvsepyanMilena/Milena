```
class Transport:
  def __init__(self, v, w, n):
    self.view=v      #вид
    self.weight=w    #вес
    self.numbers=n   #номера

class Car(Transport):
  def Carcolour(self, c):
    self.colour=c    #цвет 
    return self.colour

class Train(Transport):
  def Trainspeed(self,  s): 
    self.speed=s
    return self.speed

class Express(Train):
  def speedExpress(self, s):
    self.speed=s
    return self.speed



T1 = Car("passenger", 5000, "88f560")
print("Car is {}, weighs: {}, numbers: {}, colour: {}".format(T1.view, T1.weight, T1.numbers, T1.Carcolour("red")))
T2= Train("freight", 9200, "85h465" )
print("Train is {}, weighs: {}, numbers: {}, speed: {} km/h".format(T2.view, T2.weight, T2.numbers, T2.Trainspeed(100)))
T3=Express("Express", 10000, "65j854")
print("Train is {}, weighs: {}, numbers: {}, speed: {} km/h".format(T3.view, T3.weight, T3.numbers, T3.speedExpress(150)))

```
